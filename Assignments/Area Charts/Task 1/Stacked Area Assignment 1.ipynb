```python
#import the libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#Import the dataset
df_fuel_engine_types=pd.read_csv("//Users//motolanikay-salami//Downloads//stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
#Plot the chart
colors=["#DC143C", "#104E8B", "#0000FF", "#000000"]
labels=["Other", "Diesel", "Petrol", "Gas"]
plt.figure(figsize=(12,6))
sns.set_style("white")
plt.stackplot(df_fuel_engine_types["Year"],
             df_fuel_engine_types["Other"],
             df_fuel_engine_types["Diesel"],
             df_fuel_engine_types["Petrol"],
            df_fuel_engine_types["Gas"],
             colors=colors,
             edgecolor="None")
plt.xticks(df_fuel_engine_types["Year"], rotation=45)
plt.legend(labels=labels, loc="upper left")
plt.ylabel("Number of Cars", fontsize=13)
plt.title("Populatiry of Engine fuel Types (1982-2016)", fontsize=14, weight="bold")
sns.despine()
plt.show()
plt.savefig('area chart assignmnet 1.png')
```

```python
#import the libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#Import the dataset
df_printer_revenue=pd.read_csv("//Users//motolanikay-salami//Downloads//area_chart_homework_ii_data.csv")
```

```python
df_printer_revenue
```

```python
#Plot the chart
colors=["#DC143C", "#104E8B", "#0000FF"]
labels=["Domestic", "International", "Servies", ]
plt.figure(figsize=(12,6))
sns.set_style("white")
plt.stackplot(df_printer_revenue["Date"],
             df_printer_revenue["Domestic"],
             df_printer_revenue["International"],
             df_printer_revenue["Services"],
             colors=colors,
             edgecolor="None")
plt.xticks(df_printer_revenue["Date"], rotation=45)
plt.legend(labels=labels, loc="upper left")
plt.ylabel("Number of Revenue", fontsize=13)
plt.title("Printers Revenue for the Last 12 months", fontsize=14, weight="bold")
sns.despine()
plt.show()
plt.savefig('area chart assignmnet 2.png')
```

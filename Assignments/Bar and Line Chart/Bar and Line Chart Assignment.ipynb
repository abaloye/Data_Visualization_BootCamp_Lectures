```python
#load the packages
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
df_movie_complaints=pd.read_csv("//Users//motolanikay-salami//Downloads//bar_line_chart_homework.csv")
```

```python
df_movie_complaints
```

```python
fig,ax=plt.subplots(figsize=(14,9))

ax.bar(df_movie_complaints["Movie Theater Goers Complaints"],
       df_movie_complaints["Number of complaints"],
       color="k")
ax.set_ylabel("Number of complaints",
             weight="bold")
ax.tick_params(axis="y",
               width=2,
               labelsize="large")
ax1=ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1.0))
ax1.plot(df_movie_complaints["Movie Theater Goers Complaints"],
         df_movie_complaints["frequency"],
         color="#308014",
         marker="D")
ax1.set_ylabel("frequency",
               color="#308014",
               weight="bold")
ax1.tick_params(axis="y",
                colors="#308014",
                width=2,
                labelsize="large")
ax.set_title("Movie Theatre Goers Complaints", fontsize=14, fontweight="bold")
plt.savefig("Bar and line charts assignment.png")
```

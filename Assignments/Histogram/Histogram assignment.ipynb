```python
#load the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#Import the dataset
df_age_participants=pd.read_csv("//Users//motolanikay-salami//Downloads//histogram_survey_data.csv")
```

```python
df_age_participants
```

```python
#Plot the chart
sns.set_style("white")
plt.figure(figsize=(8,6))
plt.hist(df_age_participants["Age"],
         bins=10,
        color="#CD0000")
plt.title("Geodemographic features of Participants", fontsize=14, fontweight="bold")
plt.xlabel("Ages of Participants")
plt.ylabel("Number of Participants")
sns.despine()
plt.show()
plt.savefig("Histogram.png")
```

```python
#Load the libraries
import pandas as pd #for importing and loading the dataset
import matplotlib.pyplot as plt # for data visualization
import seaborn as sns #for data visualization
sns.set()
```

```python
#load our data
df_car_insurance=pd.read_csv("//Users//motolanikay-salami//Downloads//regression_auto_insurance_sweden.csv")
```

```python
df_car_insurance
```

```python
#Plot the chart
#using different sns styles to plot the regression model
#plt.figure(figsize= (10,8)) #increasing the chart size
#regplot accepts variety of formats for the x and y variable such as numpy and pandas series
sns.set(rc={"figure.figsize" : (9,6)}) #using sns to set figsize
sns.regplot(x="Claims",
            y="Payment",
           data=df_car_insurance,
           scatter_kws={"color":"k"},#change the colour of the points
           line_kws={"color" : "yellow"})#Change the colour of the line
plt.xlabel("Claims")
plt.ylabel("Payment")
plt.title("Relationship between Claims and Payment of Cars Insurance", fontsize=14, weight="bold")
plt.show()
```

```python
#lmplot can only support x and y in a form of strings
sns.lmplot(x="Claims",
            y="Payment",
           data=df_car_insurance,
           height=10,
          scatter_kws={"color":"k"},#change the colour of the points
           line_kws={"color" : "yellow"})#Change the colour of the line)
plt.xlabel("Claims")
plt.ylabel("Payment")
plt.title("Relationship between Claims and Payment of Cars Insurance", fontsize=14, weight="bold")
plt.show()
plt.savefig("Regression Plot Assignment.png")
```

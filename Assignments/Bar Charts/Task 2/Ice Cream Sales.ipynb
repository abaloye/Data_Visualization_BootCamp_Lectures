```python
#Import the necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#Load the dataset
ice_cream=pd.read_csv("//Users//motolanikay-salami//Downloads//bar_chart_homework_II_data.csv")
ice_cream
```

```python
#Plot the chart
plt.figure(figsize=(9,6))
plt.bar(x=ice_cream["Cities"],
        height=ice_cream["Frequency"],
       color= "magenta")
plt.xticks(rotation=45, fontsize=13)
plt.yticks(fontsize=13)
plt.title("Ice Cream Sales Over Cities", fontsize=16, fontweight="bold")
plt.ylabel("Frequency", fontsize=13)
plt.show()        
plt.savefig("Ice Cream Sales.png")
```
